```python
# Data preprocessing
# Data preprocessing tools(steps):
# 1. Importing libraries
# 2. importing dataset
# 3. Taking care of missing data(cleaning data)
# 4. Encoding categorical data
# i. Encoding independent variable
# ii. Encoding dependent variable
# 5. Splitting dataset into training and testing dataset
# 6. Feature scaling
```

```python
#1. importing libraries
import numpy as np
import pandas as pd
```

```python
#2. import dataset
df = pd.read_csv("car_data.csv")
print(df)

print("-------------------------------------------------")
print(df.info())
```

```python
# 3. Taking care of missing data(cleaning data)
#3.1 cleaning empty cells
# remove rows

df.dropna(inplace = True) #drop all null values from df
print(df.info())
```

```python
#3.2 remove wrong data
#check unique values from fuel_type column
print(df["fuel_type"].unique())# return unique values
#fuel_type column is correct data
```

```python
#check unique values from year column
print(df["year"].unique())
print(df["year"].info())
# all data is correct , convert year column datatype to integer(int32) or numeric(int64) 
# df["year"] = pd.to_numeric(df["year"])#int64
# print(df["year"].info())
df['year']=df['year'].astype(int)#int32
print(df["year"].info())
print(df["year"].unique())
#now year column is ready
```

```python
#check unique values from year column
print(df["kms_driven"].unique())

#Replace , and kms in kms_driven
df["kms_driven"] = df["kms_driven"].str.replace(',', '').str.replace(' kms', '')
#car['kms_driven']=car['kms_driven'].str.split().str.get(0).str.replace(',','')
print(df["kms_driven"].unique())

#change data type to integer
df['kms_driven'] = df['kms_driven'].astype(int)
print(df.info())
#now kms_driven column is ready
```

```python
#check unique values from Price column
print(df["Price"].unique())
print("-----------------------------------------------------------------------------")

#Remove records which contains Ask For Price 
df = df[df['Price'] != 'Ask For Price']
print(df["Price"].unique())
print("-----------------------------------------------------------------------------")

#Replace , in Price
df['Price'] = df['Price'].str.replace(",", "")
print(df["Price"].unique())
print("-----------------------------------------------------------------------------")

#change data type to integer
df['Price'] = df['Price'].astype(int)
print(df.info())
#now Price column is ready
```

```python
#check unique values from company column
print(df["company"].unique())
#now company column is ready
print("-----------------------------------------------------------------------------")
```

```python
#check unique values from name column
print(df["name"].unique())

#keep first 3 words from name
df["name"] = df["name"].str.split(' ').str.slice(start=0, stop=3).str.join(' ')
print(df["name"].unique())
```

```python
#3.3 remove duplicate data
#check if any duplicate record
print(df.duplicated().unique())
#Remove duplicated
df.drop_duplicates(inplace = True)
print(df.info())
#Now all data is ready
```

```python
#reindexing and saving cleaned dataset into another new file for further use
df = df.reset_index(drop=True)
df.to_csv("cleaneddata.csv")
```

```python
#Implementation of algo

#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read new cleaned dataset
df = pd.read_csv("cleaneddata.csv")
```

```python
#spliting dataset into dependent and independent variable
X = df[['name','company','year','kms_driven','fuel_type']]
y = df[['Price']]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']), 
                                       remainder = 'passthrough')
reg = LinearRegression()
pipe = make_pipeline(column_trans, reg)

scores = []
for i in range(0, 100):
    #Split data into training and testing
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.01, random_state = i)

    pipe.fit(X_train, y_train)

    y_pred = pipe.predict(X_test)
    score = r2_score(y_test, y_pred)
    scores.append(score)
    
minscoreindex = 0
minscoreindex = np.argmin(scores)
# minscoreindex = np.argmax(scores)
print(minscoreindex)
print(scores[minscoreindex])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.01, random_state = minscoreindex)
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
print(y_test)
print(y_pred)
```

```python
#my own input
myinput = np.array(['Tata Zest XM', 'Tata', 2018, 0, 'Diesel']).reshape(1, 5)
columns = ['name', 'company', 'year', 'kms_driven', 'fuel_type']

data = pd.DataFrame(columns = columns, data = myinput)
print(data)
result = pipe.predict(data)
print(int(result[0][0]))
```
